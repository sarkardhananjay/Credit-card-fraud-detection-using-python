```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('C:\\Users\\dhana\\Desktop\\allcsv\\creditcard.csv')
```

```python
legit = df[df.Class == 0]
fraud = df[df.Class == 1] 
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transaction
df.groupby('Class').mean()
```

```python
# Under Sampling
legit_sample = legit.sample(n=492)
```

```python
newdf = pd.concat([legit_sample, fraud], axis=0)
```

```python
newdf.head()
```

```python
newdf['Class'].value_counts()
```

```python
newdf.groupby('Class').mean()
```

```python
x = newdf.iloc[: , :-1].values
y= newdf.iloc[: , -1].values
```

```python
print(x)
```

```python
print(y.shape)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=3)
```

```python
print(x_train.shape, x_test.shape)
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression on the traning data
model.fit(x_train, y_train)
```

```python
# model evaluation
# accuracy on traning data
x_train_prediction = model.predict(x_train)
traning_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('Accuracy on traning data : ', traning_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
